## Python Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x ** 2
```

```python
y
```

```python
# Functional method to plot a graph
plt.plot(x, y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Title')
plt.show()
```

```python
# subplots
plt.subplot(1, 2, 1) # row,column, subplot number
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2) # row, columnm, subplot number
plt.plot(y, x, 'g')
```

```python
# Object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.9, 0.9]) # left, botton, width, height
axes.plot(x, y)
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Title')
```

```python
# Object oriented method
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9]) # left, botton, width, height
axes2 = fig.add_axes([0.7, 0.15, 0.2, 0.2]) # left, botton, width, height
axes1.plot(x, y)
axes1.set_xlabel('X Axis')
axes1.set_ylabel('Y Axis')
axes1.set_title('Title')
axes2.plot(y, x)
axes2.set_xlabel('X Axis')
axes2.set_ylabel('Y Axis')
axes2.set_title('Title')
```

```python
# Subplots and axes

fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(8, 2), dpi=100) # in figsize you pass a tuple to specify height and width; dpi is dots per pixel

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2)) # in figsize you pass a tuple to specify height and width; dpi is dots per pixel

axes[0].plot(x, y)

axes[1].plot(y, x)

plt.tight_layout()
```

## Plotly and Cufflinks

* Plotly is an interactive visualization librbary
* Cufflinks connects plotly with pandas

```python
import pandas as pd
import numpy as np
%matplotlib inline
from plotly import __version__
```

```python
print(__version__)
```

```python
import cufflinks as cf
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True) # connect plotly javascript to notebook
```

```python
cf.go_offline() # allows to use cufflinks offline
```

```python
df = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [32, 43, 50]})
```

```python
df2.head()
```

```python
df.plot(figsize=(12,7)) # matplotlib based
```

```python
df.iplot() # Plotly
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers', size=10)
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x': [1,2,3,4,5],
                   'y':[10,20,30,20,10],
                   'z':[5,4,3,2,1]})
```

```python
df3.iplot(kind='surface', colorscale='rdylbu')
```

```python
df.iplot(kind='hist')
```

```python
df[['A', 'B']].iplot(kind='spread')
```

```python
df.iplot(kind='bubble', x='A', y='B', size='C')
```

```python
df.scatter_matrix()
```
